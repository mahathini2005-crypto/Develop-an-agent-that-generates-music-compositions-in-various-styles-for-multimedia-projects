```python
from transformers import pipeline
import scipy.io.wavfile as wav
from IPython.display import Audio
```

```python
music_generator = pipeline("text-to-audio", model="facebook/musicgen-small")
```

```python
prompt = "sad piano music"
music = music_generator(prompt)
```

```python
sampling_rate = music["sampling_rate"]
audio_data = music["audio"]

wav.write("music.wav", rate=sampling_rate, data=audio_data)
```
